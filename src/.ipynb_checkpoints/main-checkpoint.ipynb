```python
import torch
import warnings
import numpy as np
from models.cnns import get_mobilenet_v2, get_efficientnet_b0, get_alexnet
from models.mlps import SimpleMLP, ResidualMLP
from models.llms import get_distilgpt2, get_gpt2_small
from benchmark.benchmark_cnns import benchmark_cnns
from benchmark.benchmark_mlps import benchmark_mlps
from benchmark.benchmark_llms import benchmark_llms


mlp_models = {
    "SimpleMLP": SimpleMLP(input_dim=100, hidden_dim=64, output_dim=10),
    "ResidualMLP": ResidualMLP(input_dim=100, hidden_dim=64, output_dim=10)
}

cnn_models = {
    "MobileNetV2": get_mobilenet_v2(pretrained=False),
    "EfficientNetB0": get_efficientnet_b0(pretrained=False),
    "AlexNet": get_alexnet(pretrained=False)
}

llm_models = {
    "DistilGPT-2": get_distilgpt2(),
    "GPT-2 Small": get_gpt2_small(),
    # "TinyLLaMA": get_tinyllama()  # optional, evtl. nur GPU
}

mlp_input_shape = (32, 100)
cnn_input_shape = (16, 3, 224, 224)
llm_input_text = "Hello, how are you today?"
```

```python
if not cuda_available:
    print("CUDA not available!")

# Geräte
cpu = torch.device("cpu")
cuda = torch.device("cuda")

# Anzahl Wiederholungen
n_runs = 5

# Funktion, die Benchmarks mehrfach ausführt und mittelt
def benchmark_average(benchmark_fn, *args, n_runs=5, **kwargs):
    # Ergebnisse speichern
    all_results = []
    for _ in range(n_runs):
        results = benchmark_fn(*args, **kwargs)
        all_results.append(results)
    
    # Durchschnitt berechnen
    avg_results = {}
    for name in all_results[0]:
        elapsed_list = [res[name]["elapsed"] for res in all_results]
        power_list = [res[name]["avg_power"] for res in all_results]
        energy_list = [res[name]["energy"] for res in all_results]
        avg_results[name] = {
            "avg_elapsed": np.mean(elapsed_list),
            "avg_power": np.mean(power_list),
            "avg_energy": np.mean(energy_list)
        }
    return avg_results

mlp_avg_cpu = benchmark_average(benchmark_mlps, mlp_models, input_shape=mlp_input_shape, device=cpu, n_runs=n_runs)
cnn_avg_cpu = benchmark_average(benchmark_cnns, cnn_models, input_shape=cnn_input_shape, device=cpu, n_runs=n_runs)
llm_avg_cpu = benchmark_average(benchmark_llms, llm_models, input_text=llm_input_text, device=cpu, n_runs=n_runs)

mlp_avg_gpu = benchmark_average(benchmark_mlps, mlp_models, input_shape=mlp_input_shape, device=cuda, n_runs=n_runs)
cnn_avg_gpu = benchmark_average(benchmark_cnns, cnn_models, input_shape=cnn_input_shape, device=cuda, n_runs=n_runs)
llm_avg_gpu = benchmark_average(benchmark_llms, llm_models, input_text=llm_input_text, device=cuda, n_runs=n_runs)

def print_avg_results(title, avg_results):
    print(f"\n--- {title} ---")
    for name, vals in avg_results.items():
        print(f"{name}: time={vals['avg_elapsed']:.4f}s, power={vals['avg_power']:.2f} W, energy={vals['avg_energy']:.2f} J")

print_avg_results("MLP CPU", mlp_avg_cpu)
print_avg_results("CNN CPU", cnn_avg_cpu)
print_avg_results("LLM CPU", llm_avg_cpu)
print_avg_results("MLP GPU", mlp_avg_gpu)
print_avg_results("CNN GPU", cnn_avg_gpu)
print_avg_results("LLM GPU", llm_avg_gpu)
```

```python
import matplotlib.pyplot as plt
import numpy as np

all_results = {
    "MLP": {"CPU": mlp_avg_cpu, "GPU": mlp_avg_gpu},
    "CNN": {"CPU": cnn_avg_cpu, "GPU": cnn_avg_gpu},
    "LLM": {"CPU": llm_avg_cpu, "GPU": llm_avg_gpu}
}

# Calculate latency
latency_results = {}

for model_type in all_results:
    latency_results[model_type] = {}

    for device in ["CPU", "GPU"]:
        latency_results[model_type][device] = {}

        for model_name, vals in all_results[model_type][device].items():

            # Determine batch sizes
            if model_type == "MLP":
                batch = mlp_input_shape[0]
            elif model_type == "CNN":
                batch = cnn_input_shape[0]
            else:
                batch = 1  # LLM = 1 Prompt

            latency = vals["avg_elapsed"] / batch
            latency_results[model_type][device][model_name] = latency

# Calculate throughput
throughput_results = {}

for model_type in all_results:
    throughput_results[model_type] = {}

    for device in ["CPU", "GPU"]:
        throughput_results[model_type][device] = {}

        for model_name, vals in all_results[model_type][device].items():

            # Determine batch size
            if model_type == "MLP":
                batch = mlp_input_shape[0]
            elif model_type == "CNN":
                batch = cnn_input_shape[0]
            else:
                batch = 1  # LLM

            # Throughput = inputs per second
            throughput = batch / vals["avg_elapsed"]
            throughput_results[model_type][device][model_name] = throughput            
            
# Helper function for plotting certain metric
def plot_model_comparison(model_type, cpu_data, gpu_data, metric_name, ylabel):
    models = list(cpu_data.keys())
    x = np.arange(len(models))
    bar_width = 0.35
    
    fig, ax = plt.subplots(figsize=(10, 5))
    
    cpu_vals = [cpu_data[m] for m in models]
    gpu_vals = [gpu_data[m] for m in models]

    ax.bar(x - bar_width/2, cpu_vals, bar_width, label="CPU", alpha=0.7)
    ax.bar(x + bar_width/2, gpu_vals, bar_width, label="GPU", alpha=0.9)

    ax.set_xticks(x)
    ax.set_xticklabels(models, rotation=45, ha="right")
    ax.set_ylabel(ylabel)
    ax.set_title(f"{model_type} – {metric_name} (CPU vs GPU)")
    ax.legend()
    plt.tight_layout()
    plt.show()

# Energy plots
for model_type in all_results:
    cpu = {m: all_results[model_type]["CPU"][m]["avg_energy"] for m in all_results[model_type]["CPU"]}
    gpu = {m: all_results[model_type]["GPU"][m]["avg_energy"] for m in all_results[model_type]["GPU"]}

    plot_model_comparison(model_type, cpu, gpu,
                          metric_name="Energy Consumption",
                          ylabel="Energy (Joule)")
    
print("--------------------------------------------------")

# Time plots
for model_type in all_results:
    cpu = {m: all_results[model_type]["CPU"][m]["avg_elapsed"] for m in all_results[model_type]["CPU"]}
    gpu = {m: all_results[model_type]["GPU"][m]["avg_elapsed"] for m in all_results[model_type]["GPU"]}

    plot_model_comparison(model_type, cpu, gpu,
                          metric_name="Execution Time",
                          ylabel="Time (s)")

print("--------------------------------------------------")
                          
# Latency plots
for model_type in latency_results:
    cpu = latency_results[model_type]["CPU"]
    gpu = latency_results[model_type]["GPU"]

    plot_model_comparison(model_type, cpu, gpu,
                          metric_name="Latency per Input",
                          ylabel="Latency (s)")
  
print("--------------------------------------------------")
 
# Throughput plots
for model_type in throughput_results:
    cpu = throughput_results[model_type]["CPU"]
    gpu = throughput_results[model_type]["GPU"]

    plot_model_comparison(model_type, cpu, gpu,
                          metric_name="Throughput",
                          ylabel="Inputs per Second")
```

```python
import pandas as pd

# Liste für alle Zeilen
table_rows = []

for model_type in ["MLP", "CNN", "LLM"]:  # feste Reihenfolge
    # Modelle alphabetisch sortieren
    models = sorted(all_results[model_type]["CPU"].keys())
    for model_name in models:
        for device in ["CPU", "GPU"]:  # CPU immer vor GPU
            vals = all_results[model_type][device][model_name]
            latency = latency_results[model_type][device][model_name]
            throughput = throughput_results[model_type][device][model_name]
            energy = vals["avg_energy"]
            elapsed = vals["avg_elapsed"]
            
            table_rows.append({
                "Model Type": model_type,
                "Model": model_name,
                "Device": device,
                "Energy (J)": energy,
                "Time (s)": elapsed,
                "Latency per Input (s)": latency,
                "Throughput (inputs/s)": throughput
            })

# DataFrame erstellen
df_results = pd.DataFrame(table_rows)

# Ausgabe
print(df_results.to_string(index=False))
```
